```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate product names
products = ['Product A', 'Product B', 'Product C', 'Product D']

# Simulate sales data
data = {
    'Date': pd.date_range(start='2023-01-01', end='2023-12-31', freq='D'),
    'Product': np.random.choice(products, size=365),
    'Quantity Sold': np.random.randint(1, 100, size=365),
    'Price per Unit': np.random.uniform(10, 100, size=365).round(2)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate Total Sales for each transaction
df['Total Sales'] = df['Quantity Sold'] * df['Price per Unit']

# Display the first few rows of the DataFrame
print(df.head())

# Total sales per product
total_sales_per_product = df.groupby('Product')['Total Sales'].sum().reset_index()

# Display the total sales per product
print("\nTotal Sales per Product:")
print(total_sales_per_product)

# Best-selling product
best_selling_product = total_sales_per_product.loc[total_sales_per_product['Total Sales'].idxmax()]

print("\nBest-Selling Product:")
print(best_selling_product)

# Add a 'Month' column
df['Month'] = df['Date'].dt.to_period('M')

# Monthly sales trend
monthly_sales = df.groupby('Month')['Total Sales'].sum().reset_index()

# Convert 'Month' back to a datetime object for plotting
monthly_sales['Month'] = monthly_sales['Month'].dt.to_timestamp()

print("\nMonthly Sales Trend:")
print(monthly_sales)

# Plotting the monthly sales trend
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales['Month'], monthly_sales['Total Sales'], marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```
